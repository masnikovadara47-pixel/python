# 4.1. Циклы с заданным числом повторений

```python
from tabulate import tabulate

# Создаем заголовки столбцов (от 1 до 15)
headers = [f"{i:X}" for i in range(1, 16)]

# Создаем данные для таблицы
data = []
for i in range(1, 16):
    row = []
    for j in range(1, 16):
        product = i * j
        row.append(f"{product:X}")
    data.append(row)

# Добавляем заголовки строк (первый столбец)
for i, row in enumerate(data, start=1):
    row.insert(0, f"{i:X}")

# пустой первым элементом для выравнивания
headers_with_index = [""] + headers

# Вывод таблицы
print(tabulate(data, headers=headers_with_index, tablefmt="grid"))

```

# 4.2. Суммы и произведения числовых последовательностей

```python
n = int(input("значение n: "))

# Проверяем условие n > 1
if n > 1:
    S = 0
    factorial = 1
    
    # Вычисляем сумму факториалов
    for i in range(1, n + 1):
        factorial *= i  # вычисляем i!
        S += factorial  # добавляем к сумме
    
    # Формируем строку с уравнением
    equation_parts = []
    for i in range(1, n + 1):
        equation_parts.append(f"{i}!")
    
    equation = " + ".join(equation_parts)
    print(f"S = {equation} = {S}")
else:
    print("n должно быть больше 1")
```

# 4.3. Итерационные циклы

```python
import math

eps = float(input("Введите ε: "))

# а) a_n = arctan(a_(n-1)), a_1 = 1
print("а)")
a = [1.0]
while True:
    a.append(math.atan(a[-1]))
    if abs(a[-1] - a[-2]) < eps:
        break

print("Наименьший номер n =", len(a))
for i in range(len(a)):
    print(f"  a[{i+1}] = {a[i]:.6f}")

# б) a_n = 2 + 1/a_(n-1), a_1 = 2
print("б)")
a = [2.0]
while True:
    a.append(2 + 1 / a[-1])
    if abs(a[-1] - a[-2]) < eps:
        break

print("Наименьший номер n =", len(a))
for i in range(len(a)):
    print(f"  a[{i+1}] = {a[i]:.6f}")

# в) a_n = 0.5 * tg(a_(n-1)), a_1 = 0.5
print("в)")
a = [0.5]
while True:
    a.append(0.5 * math.tan(a[-1]))
    if abs(a[-1] - a[-2]) < eps:
        break

print("Наименьший номер n =", len(a))
for i in range(len(a)):
    print(f"  a[{i+1}] = {a[i]:.6f}")
```

# 4.4 Табулирование функций

```python
import math
from tabulate import tabulate

# Ввод данных
a = float(input("Введите начало отрезка a: "))
b = float(input("Введите конец отрезка b: "))
h = float(input("Введите шаг h: "))

table = []
x = a

while x <= b + h/2:  # небольшой запас для точности float
    y = 1 / math.tan(x / 3) + 0.5 * math.sin(x)
    table.append([round(x, 2), round(y, 4)])
    x += h

print(tabulate(table, headers=["x", "F(x)"], tablefmt="grid"))
```

# 4.5. Ввод и обработка последовательностей

```python
# Ввод последовательности одной строкой
numbers = list(map(float, input("Введите числа через пробел: ").split()))

# Поиск максимума по модулю
max_abs = numbers[0]
for num in numbers:
    if abs(num) > abs(max_abs):
        max_abs = num

print(f"Максимальный по модулю элемент: {max_abs}")
print(f"Его модуль: {abs(max_abs)}")
```
